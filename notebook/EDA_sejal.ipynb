```python
def label_encode(column_list, X_train):
    le = LabelEncoder()
    labelled_cols = []
    for col in column_list:
        if len(list(X_train[col].unique())) <= 2:
            X_train[col] = le.fit_transform(X_train[col])
            labelled_cols.append(col)
    return labelled_cols
```

```python
X_train[non_numeric_cols].head()
```

```python
labelled_cols = label_encode(non_numeric_cols, X_train)
labelled_cols = label_encode(non_numeric_cols, X_test)
```

```python
def one_hot_encode(X_train, cols):
    temp = pd.get_dummies(X_train[cols])
    X_train = pd.concat([X_train,temp], axis=1)
    X_train.drop(cols, inplace=True, axis=1)
    return X_train
```

```python
X_train = one_hot_encode(X_train, list(set(non_numeric_cols) - set(labelled_cols))) 
```

```python
X_test.head()
```

```python
X_test = one_hot_encode(X_test, list(set(non_numeric_cols) - set(labelled_cols))) 
```

```python
X_test.head()
```

```python
X_train.fillna(1, inplace=True)
X_test.fillna(1, inplace=True)
```

```python
set(X_train.columns)-set(X_test.columns)
X_train.drop( "NAME_FAMILY_STATUS_Unknown", inplace=True, axis=1)
```

```python
s_scaler.fit(X_train)
X_train_scaled = s_scaler.transform(X_train)
```

```python
X_test_scaled = s_scaler.transform(X_test)
```
